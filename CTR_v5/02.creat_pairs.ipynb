```python
import pandas as pd
import numpy as np
import datetime
import pickle
from tqdm import tqdm, trange
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from os import listdir
from os.path import isfile, isdir, join

# 指定要列出所有檔案的目錄
mypath = './weblog_sorted'
# 取得所有檔案與子目錄名稱
files = listdir(mypath)
csv_list = []
# 以迴圈處理
for f in files:
  # 產生檔案的絕對路徑
  fullpath = join(mypath, f)
  # 判斷 fullpath 是檔案還是目錄
  if isfile(fullpath):
#     print("檔案：", f)
    if f[-4:]=='.csv':
        csv_list.append(f)
csv_list.sort()
```

```python
from itertools import tee, islice, zip_longest
def get_next(some_iterable, window=1):
    items, nexts = tee(some_iterable, 2)
    nexts = islice(nexts, window, None)
    return zip_longest(items, nexts)
```

```python
lt_piar_count = []
for date in csv_list:
    pid_pairs = []
    with open('./pid_corpus/' + 'pid_corpus_' + date[:-4] + '.pkl', 'rb') as file:
        pid_corpus_date = pickle.load(file)
    for cs in pid_corpus_date:
        for target, _next in get_next(cs):
            if _next is not None:
                pid_pairs.append((target, _next))
    lt_piar_count.append(len(pid_pairs))
    with open('./pid_pairs/' + 'pid_pairs_' + date[:-4] + '.pkl', 'wb') as file:
        pickle.dump(pid_pairs, file)
```

```python
lt_piar_count = []
for date in csv_list:
    pid_pairs = []
    with open('./pid_corpus/' + 'pid_corpus_' + date[:-4] + '.pkl', 'rb') as file:
        pid_corpus_date = pickle.load(file)
    for cs in pid_corpus_date:
        for target, _next in get_next(cs):
            if _next is not None:
                pid_pairs.append((target, _next))
    lt_piar_count.append(len(pid_pairs))
```

```python
superimposed = []
for date in lt_piar_count:
    if not superimposed:
        superimposed.append(date)
    else:
        superimposed.append(superimposed[-1]+date)
```
